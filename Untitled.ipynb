```python
data = pd.read_csv('data/vaccination_all_tweets.csv')
data.head()
```

```python
data1 = data.head(100)
```

```python
st = time.time()
ret = []
for idx in data1.index:
    address = data1.loc[idx, 'user_location']
    try:
        ret.append(geolocator.geocode(address).split(' ')[-1])
    except:
        ret.append(None)
ed = time.time()
print('Computation time:', (ed - st) / 60)
```
